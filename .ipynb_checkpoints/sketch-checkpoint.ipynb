```python
import cv2
import matplotlib.pyplot as plt

def show_image(title, img, is_gray=False):
    plt.figure(figsize=(6, 6))
    if is_gray:
        plt.imshow(img, cmap='gray')
    else:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.imshow(img)
    plt.title(title)
    plt.axis('off')
    plt.show()

# Read the image
img = cv2.imread("OIP.jpeg")
show_image("Original Image", img)

```

```python
# Convert to grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
show_image("Grayscale Image", img_gray, is_gray=True)

```

```python
# Invert the grayscale image
img_invert = cv2.bitwise_not(img_gray)
show_image("Inverted Image", img_invert, is_gray=True)

```

```python
# Apply Gaussian Blur
img_smoothing = cv2.GaussianBlur(img_invert, (21, 21), sigmaX=0, sigmaY=0)
show_image("Smoothed Image", img_smoothing, is_gray=True)

```

```python
# Define dodgeV2 function
def dodgeV2(x, y):
    return cv2.divide(x, 255 - y, scale=256)

# Create final sketch-like image
final_img = dodgeV2(img_gray, img_smoothing)
show_image("Final Sketch", final_img, is_gray=True)
```
